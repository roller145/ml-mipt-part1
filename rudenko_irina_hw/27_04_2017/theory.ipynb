**Вопрос 1**: Чем отличаются современные сверточные сети от сетей 5 летней давности?

В первую очередь, современные сети отличаются глубиной, примерно раз в 100.

Плюс, drop-out начали использовать как раз примерно пять лет назад, что помогает использовать одну нейронную сеть как ансамбль и предотвратить переобучение
